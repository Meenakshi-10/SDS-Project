```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
df= pd.read_excel('clean_ver.xlsx', index_col=0) 
df['year'] = [d.year for d in df.Date]
df.dropna(inplace=True)
```

```python
sns.boxplot(x='year', y='NO2', data=df, showfliers=False)
```

```python
sns.boxplot(x='year', y='SO2', data=df, showfliers=False)
```

```python
sns.boxplot(x='year', y='NO', data=df, showfliers=False)
```

```python
sns.boxplot(x='year', y='CO', data=df, showfliers=False)
```

```python
sns.boxplot(x='year', y='O3', data=df, showfliers=False)
```

```python
df['total']=df['PM2.5']+df['NO']+df['NO2']+df['NOx']+df['CO']+df['SO2']+df['O3']
```

```python
sns.boxplot(x='year', y='total', data=df, showfliers=False)
```
