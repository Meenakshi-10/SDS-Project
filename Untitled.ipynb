```python
for column in data[['PM2.5', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]:
    print(column)
    count = data[column].isna().sum()
    print((count/29463)*100)
```

```python
data = data.drop(columns = ['Xylene', 'NH3'])
data
```

```python
for column in data[['PM2.5', 'NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene']]:
    m = data[column].mean(skipna = True)
    print(column, m)
    data[column].fillna(m, inplace = True)
```
